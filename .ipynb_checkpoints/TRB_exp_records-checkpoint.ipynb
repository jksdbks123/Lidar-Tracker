```python
d = 1.4
thred_s = 0.6
N = 20
delta_thred = 1e-2
step = 1e-2
```

```python
def get_thred(ts,d ,thred_s ,N ,delta_thred ,step ):
    emp_ratio = len(ts[ts==0])/len(ts)
    if emp_ratio > 0.8:
        return 200
    
    ts = ts[ts!=0]
    flag = True                             
    for i in range(N):
        sample = np.random.choice(ts,replace=False)
        if len(ts[(ts > sample - d)&(ts < sample + d)])/len(ts) > thred_s:
            flag = False
            break
    if flag:
        return 200
    
    his_pos = []
    his_delta = []
    cur_thred = sample
    cur_step = step 
    
    while True:
        next_thred = cur_thred - cur_step
        delta = (len(ts[ts > next_thred])/len(ts) - len(ts[ts > cur_thred])/len(ts))
        his_pos.append(next_thred)
        his_delta.append(delta)
        if delta < delta_thred:
            break
        cur_thred = next_thred
        
        
    return next_thred,his_pos,his_delta
```

```python
laser_id = int(8)
azi = 40
ts = aggregated_maps[np.arange(750,1000),theta == theta_raw[laser_id],int(azi/100)].flatten()
plt.figure(figsize=(10,5))
plt.scatter(np.arange(len(ts)),ts,label = 'Distance Data Point',c = 'black',marker = 'x',s = 20)
plt.ylabel('Returned Distance Value (meter)',fontsize = 20)
plt.legend(fontsize = 15)
plt.xticks(np.arange(0,250,20),np.arange(750,1000,20),fontsize = 15)
plt.yticks(fontsize = 15)
plt.ylim(15.12,15.23)
```

```python
# %%timeit
thred,his_pos,his_delta = get_thred(ts,d,thred_s,N,delta_thred,step)
```

```python
def visualize(Td_map,Labeling_map,Thred_map):
    np.random.seed(250)
    color_map = np.random.random((100,3))
    color_map = np.concatenate([color_map,np.array([[1,1,1]])])
    Xs = []
    Ys = []
    Zs = []
    Labels = []
    for i in range(Td_map.shape[0]):
        longitudes = theta[i]*np.pi / 180
        latitudes = azimuths * np.pi / 180 
        hypotenuses = Td_map[i] * np.cos(longitudes)
        X = hypotenuses * np.sin(latitudes)
        Y = hypotenuses * np.cos(latitudes)
        Z = Td_map[i] * np.sin(longitudes)
        Valid_ind =  (Td_map[i] != 0)&(Td_map[i]<Thred_map[i]) # None zero index
        Xs.append(X[Valid_ind])
        Ys.append(Y[Valid_ind])
        Zs.append(Z[Valid_ind])
        Labels.append(Labeling_map[i][Valid_ind])

    Xs = np.concatenate(Xs)
    Ys = np.concatenate(Ys)
    Zs = np.concatenate(Zs)
    Labels = np.concatenate(Labels).astype('int')
    Colors = []
    for i in range(len(Labels)):
        Colors.append(color_map[Labels[i]])
    Colors = np.array(Colors)
    pcd = op3.geometry.PointCloud()
    XYZ = np.concatenate([Xs.reshape(-1,1),Ys.reshape(-1,1),Zs.reshape(-1,1)],axis = 1)
    pcd.points = op3.utility.Vector3dVector(XYZ)
    pcd.colors = op3.utility.Vector3dVector(Colors)
    pcd.scale
    visualizer = JVisualizer()

    visualizer.add_geometry(pcd)
    visualizer.show()
```
